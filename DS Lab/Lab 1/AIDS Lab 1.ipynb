```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('train_dataset.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df['age'].value_counts()
```

```python
sns.countplot(data=df, x='age', hue='smoking')
plt.show()
```

```python
cols = list(df.columns[4:])
cols
```

```python
i = cols[0]
sns.countplot(data=df, x=i, hue='smoking')
plt.show()
```

```python
grp = df[cols].groupby('smoking').mean()
grp
```

```python
cols = cols[:-1]
```

```python
fig, ax1 = plt.subplots(nrows=9, ncols=2, figsize=(15,35))
for i in range(0,18,2):
    sns.barplot(x= ['Not Smoking','Smoking'], y=grp.iloc[:,i//2].values, width=0.5, ax=ax1[i//2,0])
    ax1[i//2,0].set_title(cols[i])
    sns.barplot(x= ['Not Smoking','Smoking'], y=grp.iloc[:,i//2 +1].values, width=0.5, ax=ax1[i//2,1])
    ax1[i//2,1].set_title(cols[i+1])
plt.show()
```
