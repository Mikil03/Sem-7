```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
df = pd.read_csv('Reviews.csv')
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.drop(['Id'], axis=1, inplace=True)
```

```python
df['Score'].value_counts()
```

```python
sns.countplot(data=df, x='Score')
plt.ylabel('Frequency')
plt.title('Disrtibution of Score')
plt.show()
```

```python
df['sentiment'] = 0
```

```python
for i in range(df.shape[0]):
    if df['Score'].iloc[i] > 3:
        df['sentiment'].iloc[i] = 1
    else:
        df['sentiment'].iloc[i] = -1
```

```python
positive = df.loc[df['sentiment'] == 1]
pos_word = " ".join([pos for pos in positive['Text']])
pos_word[:1000]
```

```python
stop = set(STOPWORDS)
pos_cloud = WordCloud(stopwords=stop).generate(pos_word)

plt.imshow(pos_cloud)
plt.axis('off')
plt.show()
```

```python
negative = df.loc[df['sentiment'] == -1]
neg_word = " ".join([neg for neg in negative['Text']])
neg_word[:1000]
```

```python
neg_cloud = WordCloud().generate(neg_word)

plt.imshow(neg_cloud)
plt.axis('off')
plt.show()
```
