```python
import pandas as pd
```

```python
admit = pd.read_csv('admit1.csv')
admit.head()
```

```python
applied = pd.read_csv('applied1.csv')
applied.head()
```

```python
reject = pd.read_csv('reject1.csv')
reject.head()
```

```python
df = pd.concat([admit,reject,applied], axis=0)
df.head()
```

```python
df.reset_index(drop=True, inplace=True)
df
```

```python
df.info()
```

```python
cols = ['status', 'name', 'province', 'country', 'university,program', 'exp',
       'paper', 'scholarship', 'cgpa_got', 'cggpa_outof', 'GRE', 'gre_verbal',
       'gre_quants', 'gre_awa', 'applied_on', 'decision_on', 'TOEFL',
       'Reading', 'Speaking', 'Listening', 'Writing']

for i in cols:
    df[i] = df[i].str.strip()
```

```python
df['country'].value_counts()
```

```python
df.rename({'cggpa_outof':'cgpa_outof'}, axis=1, inplace=True)
```

```python
df['cgpa_outof'].str.split('/')
```

```python
for i in range(df.shape[0]):
    df['cgpa_got'].iloc[i] = df['cgpa_outof'].iloc[i].split('/')[0]
    df['cgpa_outof'].iloc[i] = df['cgpa_outof'].iloc[i].split('/')[1]
df.head()
```

```python
df['program'] = pd.NA
```

```python
df.rename({'university,program':'university'}, axis=1, inplace=True)
```

```python
df['university'].str.contains(' - ').sum()
```

```python
df['university'].iloc[10].split(' - ')
```

```python
for i in range(df.shape[0]):
    df['program'].iloc[i] = df['university'].iloc[i].split(' - ')[1]
    df['university'].iloc[i] = df['university'].iloc[i].split(' - ')[0]
df.head()
```

```python
df['exp'].iloc[15].split()
```

```python
for i in range(df.shape[0]):
    try:
        df['exp'].iloc[i] = df['exp'].iloc[i].split()[0]
    except:
        continue
df.head()
```

```python
df['exp'].value_counts()
```

```python
for i in range(df.shape[0]):
    try:
        df['paper'].iloc[i] = df['paper'].iloc[i].split()[0]
    except:
        continue
df.head()
```

```python
df['paper'].value_counts()
```

```python
df['scholarship'].value_counts()
```

```python
df['scholarship'].replace(to_replace={'None':0,'Data not collected':0}, inplace=True)
```

```python
for i in range(df.shape[0]):
    try:
        df['scholarship'].iloc[i] = df['scholarship'].iloc[i].split()[0]
    except:
        continue
df.head()
```

```python
df.info()
```

```python
for i in ['exp','paper','scholarship','cgpa_got','cgpa_outof']:
    df[i] = pd.to_numeric(df[i])
df.info()
```

```python
df.drop(['GRE','TOEFL'],axis=1, inplace=True)
```

```python
df['gre_quants'].iloc[1][:3]
```

```python
for i in range(df.shape[0]):
    try:
        df['gre_quants'].iloc[i] = df['gre_quants'].iloc[i][:3]
        df['gre_verbal'].iloc[i] = df['gre_verbal'].iloc[i][:3]
    except:
        continue
df.head()
```

```python
df['gre_awa'].iloc[0].split('AWA')
```

```python
for i in range(df.shape[0]):
    try:
        df['gre_awa'].iloc[i] = df['gre_awa'].iloc[i].split('AWA')[0]
    except:
        continue
df.head()
```

```python
df['applied_on'].iloc[0].split(': ')
```

```python
df['decision_on'].iloc[0].split(': ')
```

```python
for i in range(df.shape[0]):
    try:
        df['applied_on'].iloc[i] = df['applied_on'].iloc[i].split(': ')[1]
        df['decision_on'].iloc[i] = df['decision_on'].iloc[i].split(': ')[1]
    except:
        continue
df.head()
```

```python
df['decision_on'].iloc[1] + " " + str(int(df['Year'].iloc[1]))
```

```python
for i in range(df.shape[0]):
    try:
        df['decision_on'].iloc[i] = df['decision_on'].iloc[i] + " " + str(int(df['Year'].iloc[i]))
    except:
        continue
df.head()
```

```python
df.drop(['Year'],axis=1,inplace=True)
```

```python
for i in range(df.shape[0]):
    try:
        df['Reading'].iloc[i] = df['Reading'].iloc[i][:2]
        df['Speaking'].iloc[i] = df['Speaking'].iloc[i][:2]
        df['Listening'].iloc[i] = df['Listening'].iloc[i][:2]
        df['Writing'].iloc[i] = df['Writing'].iloc[i][:2]
    except:
        continue
df.head()
```

```python
df.to_csv('canada_decisions.csv',index=False)
```
