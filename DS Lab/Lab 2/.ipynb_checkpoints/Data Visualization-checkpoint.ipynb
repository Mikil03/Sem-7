```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('canada_decisions.csv')
df.head()
```

```python
df.info()
```

```python
# Distribution of Status
plt.figure(figsize=(12,5))

plt.suptitle('Statuswise Distribution')

plt.subplot(1,2,1)
plt.pie(df['status'].value_counts().values, autopct='%0.2f%%', labels=df['status'].value_counts().keys())

plt.subplot(1,2,2)
sns.countplot(data=df, x = 'status')
plt.ylabel('Number of Applications')
plt.xlabel('Status')

plt.show()
```

```python
plt.pie(df['country'].value_counts().values[:5], autopct='%0.2f%%', labels=df['country'].value_counts().keys()[:5], explode=[0,0,.2,.3,.4])
plt.title('Top 5 Countries')
plt.show()
```

```python
df.drop(df['cgpa_outof'])
```

```python
programs = ['Applied Computing',
 'Applied Computing (MSc)',
 'Applied Computing Technology',
 'Applied Data Analytics',
 'Big Data Analytics',
 'Big Data Analytics Stream: Applied Modeling and Quantitative Methods',
 'Computer Engineering',
 'Computer Science',
 'Computing Science MSc',
 'Data Analytics',
 'Data Science',
 'Data Science (MSc)',
 'Data Science and Analytics',
 'Electrical & Computer Engineering',
 'Electrical and Computer Engineering',
 'Electronics and Computer Engineering',
 'Engineering Electrical and Computer',
 'Machine Learning and Data Science']
```

```python
comp = df.loc[df['program'].isin(programs)]
```

```python
comp.reset_index(drop=True, inplace=True)
```

```python
# Distribution of Status
plt.figure(figsize=(12,5))

plt.suptitle('Statuswise Distribution')

plt.subplot(1,2,1)
plt.pie(comp['status'].value_counts().values, autopct='%0.2f%%', labels=comp['status'].value_counts().keys())

plt.subplot(1,2,2)
sns.countplot(data=comp, x = 'status')
plt.ylabel('Number of Applications')
plt.xlabel('Status')

plt.show()
```

```python
for i in range(comp.shape[0]):
    try:
        if comp['cgpa_outof'].iloc[i] == 4:
            comp['cgpa_outof'].iloc[i] *= 2.5
            comp['cgpa_got'].iloc[i] *= 2.5
        elif comp['cgpa_outof'].iloc[i] == 100:
            comp['cgpa_outof'].iloc[i] /= 10
            comp['cgpa_got'].iloc[i] /= 10
        elif comp['cgpa_outof'].iloc[i] == 10:
            pass
        else:
            df.drop(i, inplace=True)
    except:
        print(i)
```

```python
comp['applied_month'] = comp['applied_on'].str.slice(0,3)
comp['decision_month'] = comp['decision_on'].str.slice(0,3)
```

```python
comp
```

```python
comp[['status','decision_month','applied_month']].groupby(by=['decision_month','applied_month']).sum()
```

```python
comp[['status','decision_month','applied_month']].replace(to_replace={'ADMIT':1,'REJECT':-1}).groupby(by=['decision_month','applied_month']).sum()
```

```python
test= comp[['status','decision_month','applied_month']]
test
```

```python
test['result'] = 0
```

```python
for i in range(test.shape[0]):
    if test['status'].iloc[i] == 'ADMIT':
        test['result'].iloc[i] = 1
    elif test['status'].iloc[i] == 'REJECT':
        test['result'].iloc[i] = -1
    else:
        test['result'].iloc[i] = 0
```
