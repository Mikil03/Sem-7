### Dataset Link - https://www.kaggle.com/datasets/mohammedrizwanmalik/kidney-stones-mri-and-ct-scans

```python
import os
import keras 
from keras.models import Sequential
from keras.layers import Conv1D, MaxPooling1D, Flatten, Dense, Dropout, BatchNormalization
from PIL import Image
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('dark_background')
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
```

```python
encoder = OneHotEncoder()
encoder.fit([[0], [1]]) 

# 1 - Stone
# 0 - Normal
```

```python
data = []
paths = []
result = []

for r, d, f in os.walk(r'./Dataset/Train/Kidney_stone'):
    for file in f:
        if '.png' in file:
            paths.append(r+'/'+file)

for path in paths:
    img = Image.open(path)
    img = img.resize((500,500))
    img = np.array(img)
    img = img[0:300,50:450]
    if(img.shape == (300,400)):
        data.append(np.array(img))
        # result.append([1,1])
        result.append(encoder.transform([[1]]).toarray())
```

```python
paths = []
for r, d, f in os.walk(r"./Dataset/Train/Normal"):
    for file in f:
        if '.png' in file:
            paths.append(r+'/'+file)

for path in paths:
    img = Image.open(path)
    img = img.resize((500,500))
    img = np.array(img)
    img = img[0:300,50:450]
    if(img.shape == (300,400)):
        data.append(np.array(img))
        # result.append([0,0])
        result.append(encoder.transform([[0]]).toarray())
```

```python
data = np.array(data)
data.shape
```

```python
result = np.array(result)
result = result.reshape(-1,2)
```

```python
result.shape, data.shape
```

# Splitting the Data

```python
x_train,x_test,y_train,y_test = train_test_split(data, result, test_size=0.2, shuffle=True, random_state=0)
```

# Model Building

```python
model = Sequential()

model.add(Conv1D(32, kernel_size=2, input_shape=(300,400), padding = 'Same'))
model.add(Conv1D(32, kernel_size=2,  activation ='relu', padding = 'Same'))


model.add(BatchNormalization())
model.add(MaxPooling1D(pool_size=2))
model.add(Dropout(0.25))

model.add(Conv1D(64, kernel_size = 2, activation ='relu', padding = 'Same'))
model.add(Conv1D(64, kernel_size = 2, activation ='relu', padding = 'Same'))

model.add(BatchNormalization())
model.add(MaxPooling1D(pool_size=2, strides=2))
model.add(Dropout(0.25))

model.add(Flatten())

model.add(Dense(512, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(2, activation='softmax'))

model.compile(loss = "categorical_crossentropy", optimizer='Adamax')
print(model.summary())
```

```python
y_train.shape
```

```python
history = model.fit(x_train, y_train, epochs = 30, batch_size = 166, validation_data = (x_test, y_test))
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Test', 'Validation'], loc='upper right')
plt.show()
```

```python
def names(number):
    if number==0:
        return 'Stone present'
    else:
        return 'No stone'
```

```python
from matplotlib.pyplot import imshow
img = Image.open(r"./Dataset/Test/Normal/1.3.46.670589.33.1.63735429634639786900001.4773547424251541024.png")
x = np.array(img.resize((500,500)))
x = x[0:300,50:450]
x = x.reshape(1,300,400)
res = model.predict_on_batch(x)
classification = np.where(res == np.amax(res))[1][0]
imshow(x[0])
print(str(res[0][classification]*100) + '% Confidence This Is ' + names(classification))
```
